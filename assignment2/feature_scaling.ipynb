```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, StandardScaler, Normalizer

df = pd.read_csv("avocado_encoded.csv")
df.head()
```

```python
# Select Numerical Columns
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
X_num = df[num_cols]

num_cols
```

```python
# Min-Max Scaling
minmax = MinMaxScaler()
X_minmax = minmax.fit_transform(X_num)

X_minmax[:5]
```

```python
# Max Absolute Scaling
maxabs = MaxAbsScaler()
X_maxabs = maxabs.fit_transform(X_num)

X_maxabs[:5]
```

```python
# Z-Score Standardization
standard = StandardScaler()
X_standard = standard.fit_transform(X_num)

X_standard[:5]
```

```python
# Vector Normalization
normalizer = Normalizer()
X_normalized = normalizer.fit_transform(X_num)

X_normalized[:5]
```

```python
df[num_cols] = X_standard
df.to_csv("avocado_scaled.csv", index=False)

print("Final scaled file saved")
```
